# Exploración Inicial de Datasets

Este notebook presenta un análisis exploratorio preliminar de tres conjuntos de datos candidatos, con el objetivo de evaluar su estructura, calidad y potencial analítico.

- 1 Titanic

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_titanic = pd.read_csv('../data/train.csv')
df_titanic.head()
df_titanic.info()
df_titanic.describe()
df_titanic.isnull().sum()


## Dataset estructurado con variables categóricas y numéricas
## Presenta valores faltantes en Age, Cabin y Embarked
## Adecuado para problemas de clasificación
```

- 3 MNIST

```python
df_mnist = pd.read_csv('../data/mnist_train.csv')
df_mnist.head()
df_mnist.shape
image = df_mnist.iloc[0, 1:].values.reshape(28,28)

plt.imshow(image, cmap='gray')
plt.title(f"Label: {df_mnist.iloc[0,0]}")
plt.axis('off')
plt.show()


## Dataset no estructurado (imágenes)
## Representado como 784 píxeles por imagen
## Utilizado en clasificación de imágenes
```
